### General Properties

As we can see  there are some columns of our data like 'id','imdb_id', 'homepage','tagline','overview' and 'release_date' that are not relevant for our research. Therefore we are going to drop them.

```python
# Load data. 
df=pd.read_csv('tmdb-movies.csv')
df.columns
```

```python
#drop the columns that are not of our  interest
df.drop(['id', 'imdb_id', 'homepage','tagline','overview','release_date','keywords','budget_adj','revenue_adj' ], axis=1, inplace=True)
```

Our dataset has 10866 samples and 12 features. Important features like genres and production_companies have a | dividing the categories. This is and important issue that has to be solved.

Also the vote_count have a huge variability, it will be interesting to find relations between the number of votes and the vote average, since some movies are well know on a certain social group and have high vote_average and more artistic movies or Oscar movies are more voted and therefore probably have less vote_average.

In this case most of these samples have missing data and inconsistent data like 0 budget and 0 revenue for movies with well known actors, so we are going to drop them.

### Data Cleaning 

During this process we are going to make all the modifications that we need to clean our data

* Drop na values when is need
* Remove duplicates
* Create a function to help to work with features like cast and genres that have dificult data.
* Remove the | from production companies.

```python
df.dropna(inplace=True)
df.isnull().sum()
```

```python
df.shape
```

Now we are going to remove the | from production companies. We only want to keep the name of the Production company. In the case of cast and genres we create the function count_sep that counts the number of times that an actor or a genre appear in the data set.

```python
df['production_companies']=df['production_companies'].apply(lambda x: x.split('|')[0])
```

```python
def count_sep(feature,df):
    data_con=df[feature].str.cat(sep='|')
    data=pd.Series(data_con.split('|'))
    return data.value_counts()
```

Now we are going to look for duplicated values and remove them if necessary.

```python
df.duplicated().sum()
df.drop_duplicates(inplace=True)
df.duplicated().sum()
```

## Exploratory Data Analysis

>  Now that we are cleaned our data, we are ready to move on to exploration. In this stage of our analysis we are going to formulate and answer some questions of interest related with this data. For this aim we are going to find patterns in the data and make visualizations that will help us to get a better understanding of the data.

### Research Question 1:  What is the top 10 of production companies with highest revenue? 

To answer this question we are going to group our data set by production companies and sum up the revenues for each one. Then we will sort these values to keep the 10 with highest values of revenue. We plot this results in a bar chart and as is shown Universal Pictures, Paramount Pictures and Walt Disney Pictures are the companies at the top of the list.

```python
best_prod=df.groupby('production_companies')['revenue'].sum()
top10=best_prod.sort_values( ascending=False).head(10)
sns.set(rc={'figure.figsize':(8,5)})
ax=top10.plot(kind='bar')
ax.set_title('Top 10 Production companies with highest revenue')
ax.set_xlabel('Production companies')
ax.set_ylabel('Revenue');
```

### Research Question 2: Is there a positive relation between the profit of a movie and its vote average? 

To asses the question first we need to note that we don't have a feature that indicates profit, nevertheless we can calculate it sice __profit=revenue-budget__.
With this aim we are going to create a new variable profit and then we are goint to calculate the correlation matrix of our data.


```python
df['profit']=df['revenue']-df['budget']
cor=df.corr()
cor.style.background_gradient()
```

As we can see there is a positive relation between profit and the average of vote but this relation is very week. On the other hand we can see a more strong positive relation between the profit and both popularity and vote_count. This could be expected since normaly the movies with high profit are very commercial and therefore popular and known. Note that we are only talking about relation and we are not implying any causation.

### Research Question 3: Have the movies with a high number of votes high average?

To answer this question we will will do a scatter plot of the vote average versus the vote count. As we can see from this graph in general movies with high vote count have a tendency to have high vote average, but movies with low vote count have a huge variability in their vote average.

```python
ax=df.plot(x='vote_average',y='vote_count',kind='scatter');
ax.set_title('Vote average of movies vs their vote count')
ax.set_xlabel('Vote average')
ax.set_ylabel('Vote count');
```

### Research Question 4: What are the top 5 directors with more quantity of movies?

For this we are going to group our dataset by director and count the number of movies that directed each of them. We obtain that the top 5 are 'Woody Allen' with 42 movies, 'Clint Eastwood' 34, 'Steven Spielberg' 29, 'Martin Scorsese' 26 and 'Ridley Scott' 23. Then we will sum the number of movies of each director obtaining a total of 154 movies between them. We will use a pie chart to visualize the percentage of this total corresponding to each director. 

```python
director=df.groupby('director')['original_title'].count()
top5=director.sort_values(ascending=False).head()
## total of movies among this top 5 directors
top5.values.sum()
```

```python
#creating a pie chart with top5
labels = top5.index
sizes = top5.values
explode = (0.1, 0, 0, 0,0)  
fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
plt.tight_layout()
```

### Research Question 5: What are the top 5 directors with more revenue?

In this case we are going to do a similar process as above but adding the revenues obtained for each director.
We can see that Steven Spielberg is the winner in this category and the rest of directors in the top 5 with more movies are not in the top 5 of the revenues. 

```python
director=df.groupby('director')['revenue'].sum()
top5=director.sort_values(ascending=False).head()
plt.subplots(figsize=(8,5))
plt.bar(top5.index,top5);
plt.title('Top 5 directors attending to revenue')
plt.xlabel('Director')
plt.ylabel('Revenue');
```

### Research Question 6: What genres have more releases?

For this we are going to use the function count_sep that we created before and plot our results in a bar chart. As we can see from the plot the genre Drama was the one with more releases.

```python
gen=count_sep('genres',df)
ax=gen.plot(kind='bar');
ax.set_title('Genres vs releases')
ax.set_ylabel('Number of releases');
```

### Research Question 7: In what year were released more movies?

We are going to group by release year and count the number of movies in each of them. We obtain that the year in which were released more movies was 2014. Also as we can see from the line plot the number of movies increased by year.

```python
release=df.groupby('release_year')['original_title'].count()
release.tail()
```

```python
sns.set(rc={'figure.figsize':(8,5)})
ax=release.plot(alpha=1,kind='line');
ax.set_title('Number of movies released year by year')
ax.set_xlabel('Year released')
ax.set_ylabel('Number of movies');
```

### Research Question 8: What actors have work in more movies?

Similar as in question 6 we will use the function count_sep to see what actor have work in more movies. As we can see from the bar chart the top 5 are 'Robert De Niro', 'Samuel L. Jackson', 'Bruce Willis', 'Nicolas Cage' and 'Michael Caine'.

```python
gen=count_sep('cast',df).head()
ax=gen.plot(kind='bar');
ax.set_title('Actor vs releases')
ax.set_ylabel('Number of releases');
```

### Research question 9: What are the top 10 movies according to revenues?
As we can see from the plot the movie with more revenues was Avatar.

```python
movies=df[['original_title','revenue']].set_index('original_title')
new=movies['revenue'].sort_values(ascending=False).head(10)
ax=sns.pointplot(x=new.values,y=new.index);
ax.set_title('Top 10 movies according to revenues')
ax.set_xlabel('Revenue')
ax.set_ylabel('Movies')
```
